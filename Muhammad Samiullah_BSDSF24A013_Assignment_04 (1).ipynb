```python
traffic_logs = [
    "id:501,zone:A1,vehicle:Car,speed:62,time:08:30,violations:[None],status:Smooth",
    "id:502,zone:A1,vehicle:Bike,speed:85,time:09:10,violations:[Helmet],status:Busy",
    "id:503,zone:B2,vehicle:Bus,speed:45,time:17:25,violations:[None],status:Smooth",
    "id:504,zone:C3,vehicle:Car,speed:110,time:14:15,violations:[Overspeed],status:Congested",
    "id:505,zone:A1,vehicle:Truck,speed:40,time:18:50,violations:[None],status:Smooth"
]

data = []
for log in traffic_logs:
    parts = log.split(',')
    d = {}
    for p in parts:
        k, v = p.split(':', 1)
        d[k] = v
    d["speed"] = int(d["speed"])
    d["violations"] = d["violations"].strip("[]")
    data.append(d)

speed_zone= {}
for x in data:
    z = x["zone"]
    if z not in speed_zone:
        speed_zone[z] = []
    speed_zone[z].append(x["speed"])

averg_speed_zone = {}
for z in speed_zone:
    averg_speed_zone[z] = sum(speed_zone[z]) / len(speed_zone[z])

hours = {}
for x in data:
    h = x["time"].split(":")[0]
    if h not in hours:
        hours[h] = 0
    hours[h] += 1

peak_hr = max(hours, key=hours.get)

fast_vehicles = []
for x in data:
    if x["speed"] > 80:
        fast_vehicles.append(x)

violations = {}
for x in data:
    v = x["violations"]
    if v not in violations:
        violations[v] = 0
    violations[v] += 1

violation_zone_counter = {}
for x in data:
    z = x["zone"]
    if z not in violation_zone_counter:
        violation_zone_counter[z] = 0
    if x["violations"] != "None":
        violation_zone_counter[z] += 1

safety_index = {}
for z in violation_zone_counter:
    score = 10 - violation_zone_counter[z]
    safety_index[z] = score

summ_of_vehicle = {}
for x in data:
    v = x["vehicle"]
    if v not in summ_of_vehicle:
        summ_of_vehicle[v] = []
    summ_of_vehicle[v].append(x)

congestion = []
for x in data:
    if x["status"] == "Congested":
        congestion.append(x["zone"])
high_congestion_zones = set(congestion)

def classify(t):
    h = int(t.split(":")[0])
    if 6 <= h < 12:
        return "Morning"
    if 12 <= h < 17:
        return "Afternoon"
    if 17 <= h < 20:
        return "Evening"
    return "Night"

time_class = []
for x in data:
    time_class.append((x["id"], classify(x["time"])))

report = {}
for x in data:
    z = x["zone"]
    if z not in report:
        report[z] = {"vehicles": 0, "speeds": [], "violations": [], "types": []}
    report[z]["vehicles"] += 1
    report[z]["speeds"].append(x["speed"])
    report[z]["violations"].append(x["violations"])
    report[z]["types"].append(x["vehicle"])

final_summary = {}
for z in report:
    r = report[z]
    avg_speed = sum(r["speeds"]) / len(r["speeds"])
    common_type = max(set(r["types"]), key=r["types"].count)
    v_count = len([v for v in r["violations"] if v != "None"])
    safety = "Safe"
    if v_count >= 1:
        safety = "Moderate"
    if v_count >= 2:
        safety = "Risky"
    final_summary[z] = {
        "total_vehicles": r["vehicles"],
        "avg_speed": avg_speed,
        "violations": r["violations"],
        "common_vehicle": common_type,
        "safety": safety
    }

print("q1 structured data:", data)
print("q2 average spped zone:", averg_speed_zone)
print("q3 peak hour traffic:", peak_hr)
print("q4 speed greater thatn 80 vehicle:", fast_vehicles)
print("q5 violations counter:", violations)
print("q6 saftey index measyue:", safety_index)
print("q7 summary of vehcile:", summ_of_vehicle)
print("q8 high conjections zones:", high_congestion_zones)
print("q9 time windowss:", time_class)
print("q10 final summar or repror:", final_summary)
```
